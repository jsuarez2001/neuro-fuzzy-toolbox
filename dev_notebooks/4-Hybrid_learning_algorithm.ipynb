# Premises

```python
x = torch.rand(10, 2)
y = torch.rand(10)
```

```python
loader = data.DataLoader(data.TensorDataset(x, y), batch_size = 2)
x = loader.dataset.tensors[0]
y = loader.dataset.tensors[1]
```

```python
model = ANFIS(
    input_size = 2,
    fuzzy_rules = 3,
    outputs = 1
)
model.init_premises(x)
model.show_premises_structure()
```

```python
def apply_params(optimizer, params):
    if params != {}:
        for param_group in optimizer.param_groups:
            for key, new_value in params.items():
                if key in param_group:
                    param_group[key] = new_value
```

```python
loss_function = nn.functional.mse_loss
```

```python
optimizer = torch.optim.Adam
params = {'lr': 0.1}
```

```python
optimizer = optimizer([model._fuzzification_layer._premises])
apply_params(optimizer, params)
        
for batch_x, batch_y in loader:
    batch_y_copy = batch_y.clone().detach()
    if loader.dataset.tensors[0].dtype != loader.dataset.tensors[1].dtype:
        batch_y_copy = batch_y_copy.to(batch_x.dtype)
            
    optimizer.zero_grad()
    pred = model(batch_x)
    loss = loss_function(pred, batch_y_copy)
    loss.backward()
    optimizer.step()
    
    
```
